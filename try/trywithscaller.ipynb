```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as ptl
```

```python
data=pd.read_csv("player_date_with_no_outlayer_with_no_scaller.csv")
print(data.shape)
data.head()
```

```python
corr= data.corr()
corr
```

```python
ptl.figure(figsize=(50,50))
sns.set_context('paper',font_scale=1)
sns.heatmap(data.corr(),annot=True)
```

```python
corr
```

```python
top_feature = corr.index[abs(corr['value'])>0.5]
top_feature
```

```python
df=data[top_feature]
print(df.shape)
df.head()
```

```python
ptl.figure(figsize=(50,50))
sns.set_context('paper',font_scale=1.4)
sns.heatmap(df.corr(),annot=True)
```

```python
df.to_csv("data_with_feature_chooseing.csv")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures 
```

```python
poly=PolynomialFeatures(degree=3)
xpoly=poly.fit_transform(df.drop(['value',],axis=1))
poly.fit(xpoly,df['value'])
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(xpoly,df['value'],train_size=0.8)
```

```python
model=LinearRegression()
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(LinearRegression(),xtrain,ytrain,cv=30)
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```
