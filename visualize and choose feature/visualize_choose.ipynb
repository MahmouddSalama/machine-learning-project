```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as ptl
```

```python
data=pd.read_csv("player_date_with_no_outlayer.csv")
print(data.shape)
data.head()
```

```python
corr= data.corr()
corr
```

```python
ptl.figure(figsize=(50,50))
sns.set_context('paper',font_scale=1)
sns.heatmap(data.corr(),annot=True)
```

```python
corr
```

```python
top_feature = corr.index[abs(corr['value'])>0.5]
top_feature
```

```python
df=data[top_feature]
print(df.shape)
df.head()
```

```python
ptl.figure(figsize=(50,50))
sns.set_context('paper',font_scale=1.4)
sns.heatmap(df.corr(),annot=True)
```

```python
df.to_csv("data_with_feature_chooseing.csv")
```
