```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures 
```

```python
data=pd.read_csv("data_with_feature_chooseing.csv")
print(data.shape)
data.head()
```

```python
corr=data.corr()
corr
```

```python
top_feature = corr.index[abs(corr['value'])>0.5]
top_feature
```

```python
df=data[top_feature]
df.head()
df.corr()
```

```python
poly=PolynomialFeatures(degree=3)
xpoly=poly.fit_transform(df.drop(['value',],axis=1))
poly.fit(xpoly,df['value'])
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(xpoly,df['value'],train_size=0.8)
```

```python
model=LinearRegression()
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(LinearRegression(),xtrain,ytrain,cv=30)
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```
